```python
import rdflib
from rdflib import OWL

g = rdflib.Graph()
g.parse("ontology.ttl")

o_namespace = str(g.query("SELECT ?ont WHERE { ?ont rdf:type owl:Ontology . }").bindings[0]['ont'])
```

```python
def combine_graph(onto_file_paths):
    gl = []
    o_namespaces = []
    for onto_file_path in onto_file_paths:
        sg = rdflib.Graph()
        sg.parse(onto_file_path)
        o_namespace = sg.query("SELECT ?ont WHERE { ?ont rdf:type owl:Ontology . }").bindings[0]['ont']
        gl.append(sg)
        o_namespaces.append(o_namespace)
    g = rdflib.Graph()
    for sg in gl:
        g = g + sg
    for o_namespace in o_namespaces:
        g.remove((None, OWL.imports, o_namespace))
    return g

g = combine_graph(['ontology.ttl', 'moduleA.ttl', 'moduleB.ttl'])
```
